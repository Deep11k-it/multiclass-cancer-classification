```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import input_file_name, lit

spark = SparkSession.builder \
    .appName("DBDA-Step1-Metadata") \
    .getOrCreate()
```

```python
def process_folder(folder_path, cancer_type, subclass, cancer_flag):
    df = spark.read.format("binaryFile") \
        .option("pathGlobFilter", "*.jpg") \
        .load(folder_path)
    
    return df.select(
        input_file_name().alias("image_path"),
        lit(cancer_type).alias("cancer_type"),
        lit(subclass).alias("subclass"),
        lit(cancer_flag).alias("cancer_flag")
    )
```

```python
LUNG_COLON_PATH = f"{BASE_PATH}/Lung and Colon Cancer"

lung_aca = process_folder(
    f"{LUNG_COLON_PATH}/lung_aca",
    "Lung",
    "lung_aca",
    1
)

lung_bnt = process_folder(
    f"{LUNG_COLON_PATH}/lung_bnt",
    "Lung",
    "lung_bnt",
    0
)

lung_scc = process_folder(
    f"{LUNG_COLON_PATH}/lung_scc",
    "Lung",
    "lung_scc",
    1
)

colon_aca = process_folder(
    f"{LUNG_COLON_PATH}/colon_aca",
    "Colon",
    "colon_aca",
    1
)

colon_bnt = process_folder(
    f"{LUNG_COLON_PATH}/colon_bnt",
    "Colon",
    "colon_bnt",
    0
)
```

```python
lung_colon_df = (
    lung_aca
    .union(lung_bnt)
    .union(lung_scc)
    .union(colon_aca)
    .union(colon_bnt)
)
```

```python
pdf = final_metadata_df.select(
    "image_path",
    "cancer_flag"
).toPandas()

print(pdf.head())
```

```python
from urllib.parse import unquote

pdf["image_path"] = pdf["image_path"].apply(unquote)
```

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    pdf,
    test_size=0.2,
    stratify=pdf["cancer_flag"],
    random_state=42
)

print(train_df["cancer_flag"].value_counts())
print(val_df["cancer_flag"].value_counts())
```

```python
# Step 1: Train + Temp
train_df, temp_df = train_test_split(
    pdf,
    test_size=0.3,
    stratify=pdf["cancer_flag"],
    random_state=42
)

# Step 2: Validation + Test
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["cancer_flag"],
    random_state=42
)
```

```python
train_df["cancer_flag"] = train_df["cancer_flag"].astype(str)
val_df["cancer_flag"] = val_df["cancer_flag"].astype(str)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(
    rescale=1./255
)
```

```python
class_mode="binary"
```

```python
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col="image_path",
    y_col="cancer_flag",
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    validate_filenames=False
)

val_generator = val_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col="image_path",
    y_col="cancer_flag",
    target_size=(224, 224),
    batch_size=32,
    class_mode="binary",
    validate_filenames=False
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(224,224,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation="relu"),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation="relu"),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation="relu"),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
import math

steps_per_epoch = math.ceil(train_generator.n / train_generator.batch_size)
validation_steps = math.ceil(val_generator.n / val_generator.batch_size)

print("Train steps:", steps_per_epoch)
print("Val steps:", validation_steps)
```

```python
print("Train images:", train_generator.n)
print("Val images:", val_generator.n)
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_generator,
    validation_steps=validation_steps,
    epochs=10
)

```

```python
model.save("binary_cancer_model.keras")
```

```python
import numpy as np

val_generator.reset()
y_pred_prob = model.predict(val_generator, steps=validation_steps)
y_pred = (y_pred_prob > 0.5).astype(int).ravel()

y_true = val_generator.classes
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred, target_names=["Non-Cancer", "Cancer"]))
```

```python
from sklearn.metrics import roc_auc_score
print("ROC AUC:", roc_auc_score(y_true, y_pred_prob))
```
